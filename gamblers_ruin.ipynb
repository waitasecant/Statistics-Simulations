```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gamble(init, final, p):
    remain = init
    while remain != 0 and remain != final:
        if np.random.binomial(1,p,size=1)==1:
            remain =remain + 1
        else:
            remain =remain -1
    return remain
```

```python
gamble(20,60,0.5)
```

```python
a = list()
for i in range(100):
    result =  0
    for i in range(100):
        if gamble(20,60,0.5)==0:
            result = result
        else:
            result = result + 1
    a.append(result/100)
```

```python
print(np.mean(a))
plt.hist(a)
plt.show()
# 0.33324
# 0.33325199999999994
```
