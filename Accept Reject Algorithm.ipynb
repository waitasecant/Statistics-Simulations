```python
import matplotlib.pyplot as plt
from numpy import exp, linspace
from scipy.stats import uniform, expon
```

```python
def f(x):
    return 0.5*exp(-0.5*x)  # Target density
def g(x):
    return exp(-x)  # Candidate density
```

```python
M = 2
N = 1000000
u = uniform.rvs(size = N)
y = expon.rvs(size = N,scale=1)
```

```python
x = [y[i] for i in range(N) if u[i] < f(y[i])/M*g(y[i])]
print(f"% of acceptance {len(x)*10/N}%")
```

```python
plt.figure(figsize = (15,5))
plt.hist(x, density = True, bins = 50, rwidth= 0.9)
plt.plot(linspace(0,5,num = 10000), expon.pdf(x = linspace(0,5,num = 10000),scale = 0.5), c = "red")
plt.plot(linspace(0,5,num = 10000), expon.pdf(x = linspace(0,5,num = 10000),scale = 1), c = "green")
plt.show()
```
